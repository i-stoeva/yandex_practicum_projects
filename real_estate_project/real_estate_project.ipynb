## Предобработка данных

### Переименование столбцов

Изменим названия столбцов, которые не соответствуют базовым правилам. 
В 'cityCenters_neares' приведем буквы к нижнему регистру, для этого используем код, который приводит к нижнему регистру все названия столбцов

```python
data.columns = data.columns.str.lower()
```

Для 'parks_around3000'и 'ponds_around3000' уберем из названия цифры путем переименования столбцов

```python
data.rename(columns = {'parks_around3000': 'parks_around', 'ponds_around3000': 'ponds_around'}, inplace = True)
```

```python
data.columns
```

```python
data.info()
```

### Работа с пропусками

Еще раз посмотрим на пропущенные значения, их количество и долю

```python
data.isna().sum()
```

```python
data.isna().mean()
```

Для двух столбцов можно сразу заполнить пропуски.
- 'is_apartment' -  отсутствие признака означает, что объект продажи не является аппартаментами. Поэтому вместо пропусков укажем False
- 'balcony' -  отсутствие значения предполагает количество балконов, равное 0.

```python
data['is_apartment'] = data['is_apartment'].fillna(False) 
```

```python
data['is_apartment'].value_counts()
```

```python
data['balcony'] = data['balcony'].fillna(0) 
```

```python
data['balcony'].value_counts()
```

Заполним нулями пропуски в столбцах 'parks_around', 'parks_nearest', 'ponds_around', 'ponds_nearest'. Это данные об удаленности квартиры от парков и прудов, а также об их количестве в радиусе 3км. 
Кроме того, избавимся от пропусков в столбце с данными об удаленности от аэропортов 'airports_nearest'.
Данные заполняем нулями в связи с тем, что для выводов эти столбцы нам не потребуются.

```python
data.loc[:,'parks_around':'ponds_nearest'] = data.loc[:,'parks_around':'ponds_nearest'].fillna(0)
```

```python
data['airports_nearest'] = data['airports_nearest'].fillna(0)
```

```python
data.isna().sum()
```

Заполним пропущенное название для населенного пункта, основываясь на данных об удаленности от центра города. 
Посмотрим на объявления без населенного пункта. Для этого сначала заполним пропуски значением 'unknown', которое станет условием для среза

```python
data['locality_name'] = data['locality_name'].fillna(value='unknown')
```

```python
data.query('locality_name in "unknown"')
```

Для расстояния от центра города выведем медиану, а также минимальные и максимальные значения. Это поможет понять, какие неизвестные населенные пункты мы можем считать Санкт-Петербургом на основании их удаленности от центра города.
К сводной таблице дополнительно применим сортировку по возрастанию максимального значения, чтобы определить границы по расстоянию для Санкт-Петербурга. 

```python
data_locality = data.pivot_table(index = 'locality_name', values = 'citycenters_nearest', 
                                 aggfunc = ['median', 'min', 'max'])
```

```python
data_locality.columns = ['median', 'min', 'max']
```

```python
data_locality.sort_values(by='max')
```

Из таблицы видно, что для Санкт_Петербурга характерено расстояние до центра города в диапазоне от 181 до 29493 м.
181 м - интересное значение, но нельзя исключать, что действительно есть квартиры в самом центре. 

Расстояния для неизвестных населенных пунктов варьируются от 1322 до 41294 м. 

При этом странно, что есть населенные пункты, которые ближе к центру Санкт-Петербурга, чем сам Санкт-Петербург. 
Их можно увидеть, применив сортировку не только по максимальному, но и по минимальному расстоянию. 


```python
data_locality.sort_values(by='min')
```

Исходя из полученных данных в сводной таблице, для заполнения пропусков в названии населенного пункта будем ориентироваться на следующие границы - записи, у которых расстояние для населенного пункта не превышает 17012 м, являются Санкт-Петербургом. 
17012 м - это минимальное значение для посёлка Шушары (3я запись, где сортировка по min), который каким-то магическим образом входит в границы Санкт-Петербурга.

```python
data.loc[(data['locality_name'] == 'unknown') & (
    data['citycenters_nearest'] < 17012), 'locality_name'] = 'Санкт-Петербург'
```

```python
data.loc[data['locality_name'] == 'unknown', 'locality_name'].count()
```

У нас осталось 20 населенных пунктов с неизвестным названием. В таблице также наблюдаем количество комнат, равное 0. Для квартир-студий такие значения можно заменить на 1.

```python
data.loc[(data['rooms'] == 0) & (data['studio'] == True), 'rooms'] = 1
```

Кроме того, в таблице остались пропуски, к которым  нельзя применить логичную замену или среднее/медиану. 
Для того, чтобы на следующем шаге привести тип данных float64 к int64, заполним оставшиеся пропуски значением "-1", это позволит сохранить пропущенные значения для выявления проблем в источнике данных, при этом astype() не будет ругаться на NaN.

```python
data.fillna(-1, inplace=True)
```

```python
data.isna().sum()
```

**Вывод:**

Таблица содержит много пропущенных значений, их доли велики и поэтому удаление повлечет большую потерю данных. Мы заполнили пропуски там, где было возможно применить логичную замену. Остальные пропущенные значения заменены на "-1" - так мы сохранили их для обнаружения проблемы и сделали возможным дальнейшую работу с таблицей в части предобработки данных. 
Пропуски не заполнены средними значениями или медианой, т.к. это исказит данные по площадям, высоте потолков, количеству этажей, количеству прудов и парков, расстояниям (до центра, аэропорта и т.д.). 

Т.е. жилая площадь, площадь кухни, высота потолков - эти параметры зависят от планировки, поэтому для пропусков мы не применяем отношение жилой площади или кухни к общей площади в зависимости от количества комнат. Например, одна двухкомнатная квартира может иметь большой санузел и гардеробную, за счет чего будет уменьшена жилая площадь, а другая "двушка" будет с большой жилой площадью.

Причиной возникновения пропусков можно считать то, что часть данных внесена пользователями и при этом не все поля являются обязательными. ### Этаж квартиры

К номеру этажа 'floor' применим категоризацию на "первый", "последний" и "другой". Для этого нам также понадобятся данные из столбца с количеством этажей в доме 'floors_total'

Добавим новый столбец с категориями для этажей: 
- если номер этажа "1", то "первый"
- если номер этажа = количеству этажей в доме, то "последний"
- все этажи, которые не первые и не последние - "другое".

Также учтем, что в данных о количестве этажей в доме есть пропуски, это "-1". К ним нельзя применить логичную замену или среднее/медиану. Поэтому для таких записей добавим категорию этажа "unknown", если он не первый.

Категоризацию этажей сделаем с помощью метода where().

```python
data['floor_category']=data['floor'].where(data['floor'] == 1, 'другое')\
                .where(data['floor']!=data['floors_total'], 'последний')\
                .where((data['floors_total'] != -1) & (data['floor'] != 1), 'unknown')\
                .where(data['floor'] != 1, 'первый')
```
